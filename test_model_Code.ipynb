```python
import numpy as np
import matplotlib.pyplot as plt
# Define the angles for visualization
theta = np.linspace(0, np.pi/2, 100)
x = np.sin(theta) ** 2
y = np.cos(theta) ** 2
fig, ax = plt.subplots()
ax.plot(x, y, label='|ψ> = cos(θ)|0> + sin(θ)|1>', color='blue')
circle = plt.Circle((0.5, 0.5), 0.5, color='lightgrey', fill=False)
ax.add_artist(circle)

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.axhline(0.5, color='black',linewidth=0.5)
ax.axvline(0.5, color='black',linewidth=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax.legend()
plt.title('Hadamard Gate: superposition collapse')
plt.show()
```
