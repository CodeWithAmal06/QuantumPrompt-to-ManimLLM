````python
with open("input.jsonl", "r",encoding="utf-8") as file:
    lines = file.readlines()
    
for index,line in enumerate(lines):
    if index>=25 and index<=99:
        marker1="class "
        start1=line.find(marker1)
        if start1==-1:
            continue #marker not present in the line
        end1=start1+len(marker1)
        for i in range(end1,len(line)):
            if line[i] == '(':
                classname = line[end1:i]
                break
        insert='%%manim -qm '+classname+'\\n'

        marker="```python"
        start=line.find(marker)
        if start==-1:
            continue
        end=start+len(marker)
        lines[index]=line[:end]+'\\n'+insert+line[end:]
    

with open("input.jsonl",'w', encoding="utf-8") as file:
    file.writelines(lines)
````
