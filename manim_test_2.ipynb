```python
%%manim -qm CircleToSquare

class CircleToSquare(Scene):
    def construct(self):
        blue_circle = Circle(color=BLUE, fill_opacity=0.5)
        green_square = Square(color=GREEN, fill_opacity=0.8)
        self.play(Create(blue_circle))
        self.wait()
        
        self.play(Transform(blue_circle, green_square))
        self.wait()
```

```python
%%manim -qm HadamardGateAnimation
class HadamardGateAnimation(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=70 * DEGREES, theta=30 * DEGREES)
        axes = ThreeDAxes(x_range=[-1.5, 1.5], y_range=[-1.5, 1.5], z_range=[-1.5, 1.5])
        sphere = Sphere(radius=1, resolution=(24, 24), fill_opacity=0.15, checkerboard_colors=False)
        sphere.set_color(BLUE_E)
        
        label_0 = MathTex("|0\\rangle").next_to(axes.z_axis.get_end(), UP)
        label_1 = MathTex("|1\\rangle").next_to(axes.z_axis.get_start(), DOWN)
        self.add_fixed_orientation_mobjects(label_0, label_1)
        
        # Initial state |0> points along +z
        start_vec = np.array([0, 0, 1])
        # After Hadamard: |0> -> (|0>+|1>)/sqrt(2), Bloch vector points along +x
        end_vec = np.array([1, 0, 0])
        
        arrow = Arrow3D(start=ORIGIN, end=start_vec, color=YELLOW)
        
        self.play(Create(axes), Create(sphere))
        self.play(GrowFromCenter(arrow))
        self.wait(0.5)

        title = Text("Applying Hadamard Gate", font_size=28).to_corner(UL)
        self.add_fixed_in_frame_mobjects(title)
        self.play(Write(title))

        new_arrow = Arrow3D(start=ORIGIN, end=end_vec, color=GREEN)
        self.play(Transform(arrow, new_arrow), run_time=2)
        self.wait(1)

        state_tex = MathTex(
            "H|0\\rangle = \\frac{1}{\\sqrt{2}}(|0\\rangle + |1\\rangle)",
            font_size=32
        ).to_corner(DL)
        self.add_fixed_in_frame_mobjects(state_tex)
        self.play(Write(state_tex))
        self.wait(2)
```
